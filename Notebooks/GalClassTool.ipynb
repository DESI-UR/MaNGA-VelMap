```python
import os
import shutil
from PIL import Image
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
unsorted = '/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots'
orientation = '/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/Bad_angle'
length = "/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/Bad_axes"
mask = "/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/Bad_mask"
merger = "/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/Distrubed_morphology"
questionable = "/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/questionable"
good = "/Users/Emlilio/School/Internships/UR ASTRO/Data/Plots/2x2_plots/Good_2x2"
```

```python
img_files = []

for i in os.listdir(unsorted):
    if i.lower().endswith('.png'):
        img_files.append(i)

key_to_folder = {
    'a': orientation,
    'l': length,
    'm': mask,
    'd': merger,
    'q': questionable,
    'g': good
}

for file in img_files:
    img_path = os.path.join(unsorted, file)
    img = Image.open(img_path)
    display(img)
    key = input("Type o-q to move, or 's' to skip: ").lower()
    
    if key in key_to_folder:
        dest = key_to_folder[key]
        shutil.move(img_path, os.path.join(dest, file))
    else:
        print(f"Skipped {file}")    
```
